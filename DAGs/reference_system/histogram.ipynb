```python
import numpy as np
import matplotlib.pyplot as plt

def histogram(
    src_data,
    title: str = ''
) -> None:
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)

    ax.hist(src_data, bins=100)
    ax.set_title(title)
    ax.set_xlabel('Execution time [ns]')
    ax.set_ylabel('Frequency')
    fig.set_figheight(10)
    fig.set_figwidth(12)
    fig.show()
```

```python
import yaml

JITTER_SRC = "./reference_system_exec_jitter_multi_8core_8192.yaml"
with open(JITTER_SRC, 'r') as f:
    exec_data = yaml.safe_load(f)

for node_name, exec_list in exec_data['jitter'].items():
    histogram(exec_list, node_name)
```
