```python
import matplotlib.pyplot as plt
```

```python
import glob
import yaml
from typing import Optional, Tuple

def plot(
    dir_path: str,
    y_label: str,
    x_label: str,
    y_lim: Optional[Tuple[float, float]] = None,
) -> None:
    yamls = glob.glob(f"{dir_path}/**/*.yaml", recursive=True)
    
    x = []
    y = []
    for yaml_path in yamls:
        with open(yaml_path, 'r') as f:
            result = yaml.safe_load(f)
            x.append(result[x_label])
            y.append(result[y_label])
    
    fig, ax = plt.subplots()
    
    if y_lim:
        ax.set_ylim(y_lim[0], y_lim[1])
    ax.set_xlabel(x_label, fontsize=14)
    ax.set_ylabel(y_label, fontsize=14)
    
    ax.plot(x, y, marker='+', linestyle='')

    plt.show()
```

```python
plot('./results/run_time/', 'Run time [s]', 'Number of nodes')
plot('./results/run_time/', 'Run time [s]', 'Number of edges')
plot('./results/run_time/', 'Run time [s]', 'Number of join nodes')
plot('./results/run_time/', 'Run time [s]', 'Number of timer driven nodes')
```
